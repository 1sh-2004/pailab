```python
## tictactoe_minimax_ab.py

import math
import copy

# Global node counters
nodes_minimax = 0
nodes_ab = 0

# ---------------------------------
# Part A: Core Game Functions
# ---------------------------------

def initial_state():
    return [' ' for _ in range(9)]

def print_board(state):
    print("\n")
    for i in range(0, 9, 3):
        print(state[i], "|", state[i+1], "|", state[i+2])
        if i < 6:
            print("--+---+--")
    print("\n")

def actions(state):
    return [i for i in range(9) if state[i] == ' ']

def result(state, action, player):
    new_state = state.copy()
    new_state[action] = player
    return new_state

def player(state):
    if state.count('X') == state.count('O'):
        return 'X'
    return 'O'

def winner(state):
    win_positions = [
        [0,1,2],[3,4,5],[6,7,8],
        [0,3,6],[1,4,7],[2,5,8],
        [0,4,8],[2,4,6]
    ]
    for pos in win_positions:
        if state[pos[0]] != ' ' and state[pos[0]] == state[pos[1]] == state[pos[2]]:
            return state[pos[0]]
    return None

def terminal(state):
    return winner(state) is not None or ' ' not in state

def utility(state):
    win = winner(state)
    if win == 'X':
        return 1
    elif win == 'O':
        return -1
    return 0

# ---------------------------------
# Part B: Minimax
# ---------------------------------

def max_value(state):
    global nodes_minimax
    nodes_minimax += 1

    if terminal(state):
        return utility(state)

    v = -math.inf
    for action in actions(state):
        v = max(v, min_value(result(state, action, 'X')))
    return v

def min_value(state):
    global nodes_minimax
    nodes_minimax += 1

    if terminal(state):
        return utility(state)

    v = math.inf
    for action in actions(state):
        v = min(v, max_value(result(state, action, 'O')))
    return v

def minimax_decision(state):
    best_score = -math.inf
    best_move = None
    for action in actions(state):
        value = min_value(result(state, action, 'X'))
        if value > best_score:
            best_score = value
            best_move = action
    return best_move

# ---------------------------------
# Part C: Alpha-Beta
# ---------------------------------

def max_value_ab(state, alpha, beta):
    global nodes_ab
    nodes_ab += 1

    if terminal(state):
        return utility(state)

    v = -math.inf
    for action in actions(state):
        v = max(v, min_value_ab(result(state, action, 'X'), alpha, beta))
        if v >= beta:
            return v
        alpha = max(alpha, v)
    return v

def min_value_ab(state, alpha, beta):
    global nodes_ab
    nodes_ab += 1

    if terminal(state):
        return utility(state)

    v = math.inf
    for action in actions(state):
        v = min(v, max_value_ab(result(state, action, 'O'), alpha, beta))
        if v <= alpha:
            return v
        beta = min(beta, v)
    return v

def alphabeta_decision(state):
    best_score = -math.inf
    best_move = None
    alpha = -math.inf
    beta = math.inf

    for action in actions(state):
        value = min_value_ab(result(state, action, 'X'), alpha, beta)
        if value > best_score:
            best_score = value
            best_move = action
        alpha = max(alpha, best_score)

    return best_move

# ---------------------------------
# Game Mode: Human vs AI
# ---------------------------------

def play_game():
    global nodes_minimax, nodes_ab
    state = initial_state()

    while not terminal(state):
        print_board(state)

        if player(state) == 'X':
            nodes_minimax = 0
            move = alphabeta_decision(state)
            print("AI chooses:", move)
            state = result(state, move, 'X')
        else:
            move = int(input("Enter your move (0-8): "))
            state = result(state, move, 'O')

    print_board(state)

    if winner(state):
        print(winner(state), "wins!")
    else:
        print("Draw!")

    print("\nNodes evaluated using Minimax:", nodes_minimax)
    print("Nodes evaluated using Alpha-Beta:", nodes_ab)


if __name__ == "__main__":
    play_game()
```

```python
# car_overtaking_minimax_ab.py

import math

ROAD_END = 6

nodes_minimax = 0
nodes_ab = 0

# ---------------------------------
# Part A: Problem Formulation
# ---------------------------------

def initial_state():
    return (0, 1)  # (X_position, O_position)

def actions(state, player):
    x, o = state
    moves = []
    pos = x if player == 'X' else o
    opp = o if player == 'X' else x

    for step in [1, 2]:
        new_pos = pos + step
        if new_pos <= ROAD_END and new_pos != opp:
            moves.append(step)

    return moves

def result(state, move, player):
    x, o = state
    if player == 'X':
        return (x + move, o)
    else:
        return (x, o + move)

def terminal(state):
    x, o = state
    return x == ROAD_END or o == ROAD_END

def utility(state):
    x, o = state
    if x == ROAD_END:
        return 1
    if o == ROAD_END:
        return -1
    return 0

# ---------------------------------
# Minimax
# ---------------------------------

def max_value(state):
    global nodes_minimax
    nodes_minimax += 1

    if terminal(state):
        return utility(state)

    v = -math.inf
    for move in actions(state, 'X'):
        v = max(v, min_value(result(state, move, 'X')))
    return v

def min_value(state):
    global nodes_minimax
    nodes_minimax += 1

    if terminal(state):
        return utility(state)

    v = math.inf
    for move in actions(state, 'O'):
        v = min(v, max_value(result(state, move, 'O')))
    return v

def minimax_decision(state):
    best_value = -math.inf
    best_move = None

    for move in actions(state, 'X'):
        value = min_value(result(state, move, 'X'))
        if value > best_value:
            best_value = value
            best_move = move

    return best_move

# ---------------------------------
# Alpha-Beta
# ---------------------------------

def max_value_ab(state, alpha, beta):
    global nodes_ab
    nodes_ab += 1

    if terminal(state):
        return utility(state)

    v = -math.inf
    for move in actions(state, 'X'):
        v = max(v, min_value_ab(result(state, move, 'X'), alpha, beta))
        if v >= beta:
            return v
        alpha = max(alpha, v)

    return v

def min_value_ab(state, alpha, beta):
    global nodes_ab
    nodes_ab += 1

    if terminal(state):
        return utility(state)

    v = math.inf
    for move in actions(state, 'O'):
        v = min(v, max_value_ab(result(state, move, 'O'), alpha, beta))
        if v <= alpha:
            return v
        beta = min(beta, v)

    return v

def alphabeta_decision(state):
    best_value = -math.inf
    best_move = None
    alpha = -math.inf
    beta = math.inf

    for move in actions(state, 'X'):
        value = min_value_ab(result(state, move, 'X'), alpha, beta)
        if value > best_value:
            best_value = value
            best_move = move
        alpha = max(alpha, best_value)

    return best_move

# ---------------------------------
# Game Execution
# ---------------------------------

def play_game():
    state = initial_state()

    while not terminal(state):
        print("Current State (X, O):", state)

        move = alphabeta_decision(state)
        print("X chooses move:", move)
        state = result(state, move, 'X')

        if terminal(state):
            break

        o_move = minimax_decision(state)
        print("O chooses move:", o_move)
        state = result(state, o_move, 'O')

    print("Final State:", state)

    if utility(state) == 1:
        print("X wins!")
    else:
        print("O wins!")

    print("\nNodes evaluated using Minimax:", nodes_minimax)
    print("Nodes evaluated using Alpha-Beta:", nodes_ab)


if __name__ == "__main__":
    play_game()
```
