1) Rule-Based Logical Agent for Smart Lighting
A smart building uses sensors to control lighting automatically. The system must decide
whether to turn lights ON or OFF based on logical rules.
The system perceives:
• MotionDetected
• IsDark
• ManualOverride
Represent the following knowledge using propositional logic:
• If it is dark and motion is detected, turn the light ON
• If there is no motion, turn the light OFF
• If manual override is ON, ignore all automatic rules
Design a logical agent that:
• Takes sensor readings as percepts
• Updates its knowledge base
• Infers the correct action
Given a set of percepts, determine the agent’s action using logical inference.


```python
class SmartLightingAgent:
    def decide(self, motion, is_dark, manual_override):
        if manual_override:
            return "Manual Control (Ignore automatic rules)"
        elif is_dark and motion:
            return "Light ON"
        elif not motion:
            return "Light OFF"
        else:
            return "No Change"


# Example
agent = SmartLightingAgent()
print(agent.decide(motion=True, is_dark=True, manual_override=False))
print(agent.decide(motion=False, is_dark=False, manual_override=False))
print(agent.decide(motion=True, is_dark=True, manual_override=True))
```
