5) Autonomous Warehouse Robot Using Logical Reasoning
An autonomous warehouse robot must decide how to move and operate safely using
logical reasoning rather than learning.
Percepts:
• ObstacleAhead
• PathClear
• CarryingFragileItem
• HumanNearby
• DestinationReachedRepresent operational rules using propositional logic:
• ObstacleAhead → Stop
• CarryingFragileItem ∧ ObstacleAhead → SlowDown
• HumanNearby → ReduceSpeed
• PathClear ∧ ¬HumanNearby → MoveForward
Design a logical agent architecture showing:
• Knowledge base
• Inference engine
• Action selection

```python
class WarehouseRobot:
    def decide(self, obstacle, path_clear, fragile, human_nearby):
        # Priority-based decision
        if obstacle:
            return "STOP"
        elif fragile and obstacle:
            return "SLOW DOWN"
        elif human_nearby:
            return "REDUCE SPEED"
        elif path_clear and not human_nearby:
            return "MOVE FORWARD"
        else:
            return "WAIT"


# Example
robot = WarehouseRobot()

print(robot.decide(obstacle=True, path_clear=False, fragile=False, human_nearby=False))
print(robot.decide(obstacle=False, path_clear=True, fragile=False, human_nearby=False))
print(robot.decide(obstacle=False, path_clear=True, fragile=False, human_nearby=True))
```
