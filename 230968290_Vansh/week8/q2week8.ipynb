2) Medical Diagnosis Logical Agent
A medical decision-support system assists doctors by reasoning over symptoms to
suggest possible conditions.
Given symptoms:
• Fever
• Cough
• ChestPain
Diseases:
• Flu
• Pneumonia
Represent the following medical knowledge using propositional logic:
• Fever ∧ Cough → Flu
• Fever ∧ Cough ∧ ChestPain → Pneumonia
• Pneumonia → HospitalizationRequired
Build a knowledge base and use forward chaining to infer:
• Possible diseases
• Required actions
Explain how this agent differs from a machine-learning classifier.


```python
class MedicalAgent:
    def diagnose(self, fever, cough, chest_pain):
        kb = {
            "Flu": False,
            "Pneumonia": False,
            "HospitalizationRequired": False
        }

        # Forward chaining
        if fever and cough:
            kb["Flu"] = True

        if fever and cough and chest_pain:
            kb["Pneumonia"] = True

        if kb["Pneumonia"]:
            kb["HospitalizationRequired"] = True

        return kb


# Example
agent = MedicalAgent()
result = agent.diagnose(fever=True, cough=True, chest_pain=True)

print("Diagnosis Results:")
for k, v in result.items():
    print(f"{k}: {v}")
```
