4) Logical Agent for Cybersecurity Intrusion Detection
An enterprise network uses a logical agent to detect potential cyber intrusions based on
system events.
Percepts include:
• MultipleFailedLogins
• LoginFromUnknownIP
• AccessToSensitiveFiles
• AdminPrivileges
Encode security knowledge using propositional logic:
• MultipleFailedLogins ∧ LoginFromUnknownIP → SuspiciousLogin
• SuspiciousLogin ∧ AdminPrivileges → HighRiskIntrusion
• HighRiskIntrusion → TriggerAlert
Design a knowledge-based agent that:
• Continuously updates facts
• Infers threat levels
• Decides defensive actions

```python
class CyberSecurityAgent:
    def detect(self, failed_logins, unknown_ip, sensitive_access, admin_privileges):
        kb = {
            "SuspiciousLogin": False,
            "HighRiskIntrusion": False,
            "TriggerAlert": False
        }

        # Rule 1
        if failed_logins and unknown_ip:
            kb["SuspiciousLogin"] = True

        # Rule 2
        if kb["SuspiciousLogin"] and admin_privileges:
            kb["HighRiskIntrusion"] = True

        # Rule 3
        if kb["HighRiskIntrusion"]:
            kb["TriggerAlert"] = True

        return kb


# Example
agent = CyberSecurityAgent()
result = agent.detect(True, True, True, True)

print("Cybersecurity Status:")
for k, v in result.items():
    print(f"{k}: {v}")
```
