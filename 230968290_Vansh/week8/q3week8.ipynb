3) Logical Agent for Traffic Violation Detection
A smart traffic monitoring system detects violations using logical rules rather than image-
based ML models.
Percepts include:•
SpeedAboveLimit
• SignalRed
• VehicleCrossedLine
• EmergencyVehicle
Define logical rules to infer:
• SpeedViolation
• SignalViolation
• Exceptions for emergency vehicles
Implement a logical agent cycle:
• Perceive
• Update knowledge base
• Infer violations
• Decide enforcement action
Given a sequence of percepts over time, show how conclusions change.

```python
class TrafficAgent:
    def check_violation(self, speed, red_signal, crossed_line, emergency):
        result = {
            "SpeedViolation": False,
            "SignalViolation": False
        }

        if not emergency:
            if speed:
                result["SpeedViolation"] = True
            if red_signal and crossed_line:
                result["SignalViolation"] = True

        return result


# Example sequence over time
agent = TrafficAgent()

print("Time t1:", agent.check_violation(True, False, False, False))
print("Time t2:", agent.check_violation(False, True, True, False))
print("Time t3:", agent.check_violation(True, False, False, True))
```
