```python
from collections import deque
def bfs(graph,start,goal):
    queue=deque([[start]])
    visited= set()
    while queue:
        path=queue.popleft()
        node=path[-1]
        if node==goal:
            return path
        if node not in visited:
            visited.add(node)
            for neighbor in graph[node]:
                queue.append(path+[neighbor])
            
```

```python
def dfs(graph, start, goal):
    stack = [[start]]
    visited = set()

    while stack:
        path = stack.pop()
        node = path[-1]

        if node == goal:
            return path

        if node not in visited:
            visited.add(node)

            for neighbor in graph[node]:
                if neighbor not in visited:
                    stack.append(path + [neighbor])

    return None


graph = {
    'A': ['B'],
    'B': ['A', 'C', 'D'],
    'C': ['B'],
    'D': ['B']
}

print(dfs(graph, 'A', 'D'))
```
