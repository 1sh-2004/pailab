Experiment 1: Smart Thermostat Agent (Reactive Agent)
Problem Statement: Design and implement a Smart Thermostat Agent for a single-room environment.
The room temperature can be in one of three states:
• Low
• Comfortable
• High
The agent perceives the current temperature state and can perform one of the following actions:
• Heat – increases the temperature by one level
• Cool – decreases the temperature by one level
• Idle – no action
Task Requirements
1. Implement a simple reflex agent that maintains the room at a comfortable temperature.
2. Define appropriate condition–action rules.
3. Simulate the agent’s behavior over multiple time steps starting from different initial
temperatures.
Environment Properties
• Fully observable
• Deterministic
• Sequential
• Static
• Single-agent

```python

class SmartThermostatAgent:
    def __init__(self):
        self.rules = {
            "Low": "Heat",
            "High": "Cool",
            "Comfortable": "Idle"
        }

    def perceive_and_act(self, temperature):
        return self.rules[temperature]


class RoomEnvironment:
    def __init__(self, initial_temperature):
        self.temperature_levels = ["Low", "Comfortable", "High"]
        self.current_index = self.temperature_levels.index(initial_temperature)

    def apply_action(self, action):
        if action == "Heat" and self.current_index < 2:
            self.current_index += 1
        elif action == "Cool" and self.current_index > 0:
            self.current_index -= 1

    def get_temperature(self):
        return self.temperature_levels[self.current_index]


def simulate(initial_temperature, steps=5):
    agent = SmartThermostatAgent()
    environment = RoomEnvironment(initial_temperature)

    print(f"\n initial temperature: {initial_temperature}")
    print("-" * 45)

    for step in range(steps):
        current_temp = environment.get_temperature()
        action = agent.perceive_and_act(current_temp)

        print(f"step {step + 1} temperature = {current_temp}, action = {action}")

        environment.apply_action(action)


simulate("Low")
simulate("High")
simulate("Comfortable")
```

Problem Statement: An elevator operates in a 3-floor building (Floors 1, 2, 3).
The agent perceives:
• Its current floor
• A request button pressed at one of the floors
The agent can perform:
• Move Up
• Move Down
• Stay
Task Requirements
1. Design a goal-based agent that services elevator requests.
2. Define the agent’s goal formally.
3. Implement logic that moves the elevator toward the requested floor.
4. Simulate multiple request scenarios.
Environment Properties
• Fully observable
• Deterministic
• Sequential
• Dynamic (requests may change)

```python

class ElevatorAgent:
    def decide_action(self, current_floor, requested_floor):
        if current_floor < requested_floor:
            return "Move Up"
        elif current_floor > requested_floor:
            return "Move Down"
        else:
            return "Stay"


class ElevatorEnvironment:
    def __init__(self, start_floor):
        self.current_floor = start_floor

    def apply_action(self, action):
        if action == "Move Up" and self.current_floor < 3:
            self.current_floor += 1
        elif action == "Move Down" and self.current_floor > 1:
            self.current_floor -= 1

    def get_floor(self):
        return self.current_floor


def simulate(start_floor, requested_floor):
    agent = ElevatorAgent()
    environment = ElevatorEnvironment(start_floor)

    print(f"\n Start Floor = {start_floor}, Requested Floor = {requested_floor}")
    print("-" * 55)

    step = 1
    while True:
        current_floor = environment.get_floor()
        action = agent.decide_action(current_floor, requested_floor)

        print(f"step {step}: Elevator at Floor {current_floor}, Action = {action}")

        if action == "Stay":
            print("Goal Achieved: Request Serviced")
            break

        environment.apply_action(action)
        step += 1


simulate(start_floor=1, requested_floor=3)
simulate(start_floor=3, requested_floor=1)
simulate(start_floor=2, requested_floor=2)
```

Experiment 3: Autonomous Traffic Signal Agent
Problem Statement: Design an Autonomous Traffic Signal Agent for a two-road intersection.
Each road can have:
• Low traffic
• High traffic
The agent perceives traffic density on both roads and controls the signal with actions:
• Green for Road A
• Green for Road B
Task Requirements
1. Define a PEAS description for the traffic signal agent.
2. Implement a utility-based agent that minimizes total waiting time.
3. Assign numerical utility values to traffic conditions and justify them.
4. Simulate multiple traffic scenarios.
Environment Properties
• Partially observable
• Stochastic
• Sequential
• Multi-agent (implicit)

```python
class TrafficSignalAgent:
    def __init__(self):
        self.utility = {
            "Low": 1,
            "High": 5
        }

    def decide_action(self, traffic_A, traffic_B):
        utility_A = self.utility[traffic_A]
        utility_B = self.utility[traffic_B]

        if utility_A > utility_B:
            return "Green for Road A"
        elif utility_B > utility_A:
            return "Green for Road B"
        else:
            return "Green for Road A" 


class TrafficEnvironment:
    def __init__(self, traffic_A, traffic_B):
        self.traffic_A = traffic_A
        self.traffic_B = traffic_B

    def get_state(self):
        return self.traffic_A, self.traffic_B


def simulate(traffic_A, traffic_B):
    agent = TrafficSignalAgent()
    environment = TrafficEnvironment(traffic_A, traffic_B)

    print(f"\nTraffic Scenario: Road A = {traffic_A}, Road B = {traffic_B}")
    print("-" * 55)

    state_A, state_B = environment.get_state()
    action = agent.decide_action(state_A, state_B)

    print(f"Agent Action: {action}")


simulate("Low", "Low")
simulate("High", "Low")
simulate("Low", "High")
simulate("High", "High")
```

Experiment 4: Warehouse Robot Navigation Agent
Problem Statement: A warehouse robot operates in a grid-based warehouse with fixed obstacles.
The robot perceives:
• Its current location
• Whether the current cell is blocked or free
The robot can:
• Move Up, Down, Left, Right
• Pick item at a target location
Task Requirements
1. Classify the environment using AIMA environment properties.
2. Design a model-based agent that remembers visited locations.
3. Explain why a simple reflex agent is insufficient.
4. Simulate navigation from start to target location.
Environment Properties
• Partially observable
• Deterministic
• Sequential
• Static

```python

class WarehouseRobotAgent:
    def __init__(self, start, target, grid):
        self.position = start
        self.target = target
        self.grid = grid
        self.visited = set()
        self.path = []

    def get_possible_moves(self, position):
        x, y = position
        moves = {
            "Up": (x - 1, y),
            "Down": (x + 1, y),
            "Left": (x, y - 1),
            "Right": (x, y + 1)
        }

        valid_moves = []
        for move, (nx, ny) in moves.items():
            if 0 <= nx < len(self.grid) and 0 <= ny < len(self.grid[0]):
                if self.grid[nx][ny] == 0:
                    valid_moves.append((move, (nx, ny)))
        return valid_moves

    def navigate(self):
        stack = [(self.position, [])]

        while stack:
            current, path = stack.pop()

            if current in self.visited:
                continue

            self.visited.add(current)

            if current == self.target:
                self.path = path + [current]
                return True

            for move, next_pos in self.get_possible_moves(current):
                if next_pos not in self.visited:
                    stack.append((next_pos, path + [current]))

        return False


warehouse_grid = [
    [0, 0, 1, 0],
    [0, 0, 1, 0],
    [1, 0, 0, 0],
    [0, 0, 0, 0]
]

start_location = (0, 0)
target_location = (3, 3)

robot = WarehouseRobotAgent(start_location, target_location, warehouse_grid)

print("Starting Warehouse Navigation Simulation")
print("-" * 50)

if robot.navigate():
    print("Target reached successfully!")
    print("Path followed:")
    for step in robot.path:
        print(step)
else:
    print("Failed to reach the target.")
```

Experiment 5: Online Recommendation Agent (Learning Agent)
Problem Statement: Design an Online Recommendation Agent that suggests products to users.
The agent perceives:
• User feedback (like / dislike)
The agent can:
• Recommend Product A or Product B
Task Requirements
1. Identify the four components of a learning agent:
o Performance element
o Learning element
o Critic
o Problem generator
2. Implement a simple learning mechanism that improves recommendations over time.
3. Simulate interaction over multiple user sessions.
Environment Properties
• Partially observable
• Stochastic
• Sequential
• Dynamic

```python
import random


class RecommendationAgent:
    def __init__(self, exploration_rate=0.2):
        self.scores = {
            "A": 0,
            "B": 0
        }
        self.exploration_rate = exploration_rate

    def recommend(self):
        if random.random() < self.exploration_rate:
            return random.choice(["A", "B"])
        return "A" if self.scores["A"] >= self.scores["B"] else "B"

    # Learning Element
    def learn(self, product, feedback):
        if feedback == "like":
            self.scores[product] += 1
        else:
            self.scores[product] -= 1


def user_feedback(product):
    if product == "A":
        return "like" if random.random() < 0.7 else "dislike"
    else:
        return "like" if random.random() < 0.3 else "dislike"


def simulate_sessions(num_sessions=10):
    agent = RecommendationAgent()

    print("Online Recommendation Simulation")
    print("-" * 50)

    for session in range(1, num_sessions + 1):
        product = agent.recommend()
        feedback = user_feedback(product)
        agent.learn(product, feedback)

        print(
            f"Session {session}: "
            f"Recommended Product {product}, "
            f"User Feedback = {feedback}, "
            f"Scores = {agent.scores}"
        )


simulate_sessions(15)
```
